```python
import pandas as pd
dataset = pd.read_csv('iris.csv').values
print(dataset)
```

```python
data=dataset[:,0:2]
traget=dataset[:,2]
print("data ",data)
print("traget " , traget)
```

```python
target=dataset[:,2]
print(target,target.shape)
```

```python
from matplotlib import pyplot as plt



x = data[:, 0]  
y = data[:, 1]  


for i in range(len(data)):
    if (target[i] == 0):
        
        plt.scatter(x[i], y[i], c="r", label='Class 0' if i == 0 else "")
    elif (target[i] == 1):

        
        plt.scatter(x[i], y[i], c="g", label='Class 1' if i == 0 else "")


plt.xlabel('Feature 01 (x)')
plt.ylabel('Feature 02 (y)')


plt.legend()


plt.show()
```

```python
from sklearn.model_selection import train_test_split




train_data, test_data, train_target, test_target = train_test_split(data, traget, test_size=0.1)

print("Shape of train_data:", train_data.shape)
print("Shape of test_data:", test_data.shape)
print("Shape of train_target:", train_target.shape)
print("Shape of test_target:", test_target.shape)
train_target[:5]
```
