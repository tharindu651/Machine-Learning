```python
import pandas as pd

#dataset=pd.read_csv(r'C:\Users\Migara Liyanage\Desktop\ML 5 Day Workshop 4\MLW-Day-01-main\iris.csv')
dataset=pd.read_csv('iris.csv').values

print(dataset,dataset.shape)
```

```python
data=dataset[:,0:4]
print(data,data.shape)
```

```python
target=dataset[:,4]
print(target,target.shape)
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
```

```python
print(train_data)
```

```python
print(train_target)
```

```python
print(test_data)
```

```python
print(test_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier()
#loading the KNN Algorithm ->model
```

```python
# training the algorithm
model.fit(train_data,train_target)
```

```python
#evaluating/testing
predicted_target=model.predict(test_data)
```

```python
print('actual target:',test_target)
print('predicted target:',predicted_target)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicted_target)
print('accuracy:',acc)
```
