```python
import pandas as pd
dataset =pd.read_csv('linear-insperable-dataset.csv').values
print(dataset)
```

```python
data=dataset[:,0:2]
traget=dataset[:,2]
print("data ",data)
print("traget " , traget)
```

```python
from matplotlib import pyplot as plt

x = data[:, 0]
y = data[:, 1]

for i in range(len(data)):
    if (traget[i] == 0):
        plt.scatter(x[i], y[i], c='r')#, label='Class 0' if i == 0 else "")
    elif (traget[i] == 1):
        plt.scatter(x[i], y[i], c='g')#, label='Class 1' if i == 0 else "")

plt.xlabel('Feature 01 (x)')
plt.ylabel('Feature 02 (y)')

#plt.legend()

plt.show()
```

```python
from sklearn.model_selection import train_test_split




train_data, test_data, train_target, test_target = train_test_split(data, traget, test_size=0.1)

print("Shape of train_data:", train_data.shape)
print("Shape of test_data:", test_data.shape)
print("Shape of train_target:", train_target.shape)
print("Shape of test_target:", test_target.shape)
```

```python
#from sklearn.svm import SVC
#model=SVC(kernel ='liner')
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(train_data,train_target)
```

```python
predicted_target=model.predict(test_data)
from sklearn.metrics import accuracy_score
acc=accuracy_score(test_target,predicted_target)
print('Accuracy',acc)
```
